# Carga del Dataset Anscombe

En esta celda, cargaremos el dataset `anscombe` utilizando la librería **Seaborn**. Este conjunto de datos contiene cuatro subconjuntos con propiedades estadísticas similares pero que, al ser visualizados, muestran comportamientos diferentes.

```python
# Importamos las librerías necesarias
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Cargamos el dataset desde un archivo CSV
df_anscombe = pd.read_csv("../data/df_anscombe.csv")

# Mostramos las primeras filas del dataset
df_anscombe.head()
```

# Resúmenes Estadísticos de los Conjuntos de Datos

A continuación, generaremos los resúmenes estadísticos de los cuatro subconjuntos del dataset, agrupando por el identificador `dataset`. Esto incluye medidas como la media, desviación estándar, mínimos y máximos.

```python
# Resumen estadístico por cada grupo de datos
resumen_estadistico = df_anscombe.groupby('group').describe()
resumen_estadistico
```

# Gráficas de Dispersión de los Conjuntos de Datos

Generaremos una gráfica de dispersión (scatter plot) para cada uno de los subconjuntos, utilizando diferentes colores para visualizar claramente las diferencias en su comportamiento.

```python
# Graficamos los conjuntos de datos con scatter plots
sns.relplot(
    data=df_anscombe, 
    x="x", y="y", 
    col="group", hue="group",
    kind="scatter", col_wrap=2, height=4
)

plt.suptitle("Gráficas de Dispersión de los Conjuntos de Datos", y=1.05)
plt.show()
```

# Boxplots de los Valores por Conjunto de Datos

Los **boxplots** nos permiten visualizar la distribución de los valores `x` y `y` en cada conjunto de datos. A continuación, generaremos un boxplot para `x` y otro para `y`, agrupados por el identificador del conjunto.

```python
# Boxplot para los valores de x
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_anscombe, x="group", y="x")
plt.title("Boxplot de los Valores de X por Conjunto de Datos")
plt.show()

# Boxplot para los valores de y
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_anscombe, x="group", y="y")
plt.title("Boxplot de los Valores de Y por Conjunto de Datos")
plt.show()
```

# Conclusiones

Este análisis del **dataset de Anscombe** nos muestra que, aunque los cuatro subconjuntos tienen medidas estadísticas similares (media, varianza, y correlación), al graficarlos podemos observar comportamientos muy diferentes. Esto resalta la importancia de visualizar los datos además de calcular sus estadísticas.
